```python
#This dataset does not have an ID for each occurence and as such we need to create a new column named id
df["id"] = df.index + 1
df = df[['id', 'ï»¿id_park', 'name', 'max_capacity', 'occupancy','position','datetime','entity_ts']]
df.head()
```

```python
#There is a degree of corruption in the id_park column so we need to clean its name
df.rename(columns={'ï»¿id_park': 'id_park'}, inplace=True)
df.head()
```

```python
#In 'name', there is an issue with special characters in several parking lot names, as seen in id 1

##In order to reduce the number of strings that need to be changed, we first put the entire column in lowercase
df["name"] = df["name"].str.lower()
##We then specify instances where we want to change special characters using replace
df["name"] = df["name"].str.replace("£", "").str.replace('ç', 'c').str.replace('ã', 'a').str.replace('á', 'a').str.replace('é', 'e')\
                       .str .replace('í', 'i').str.replace('ó', 'o').str.replace('ú', 'u').str.replace('â', 'a').str.replace('ê', 'e')\
                       .str.replace('î', 'i').str.replace('ô', 'o').str.replace('õ', 'o').str.replace('¢', '').str.replace('a©','e')\
                       .str.replace('a§','c')


df.head()
```

```python
df = df[['id', 'id_park', 'name', 'max_capacity', 'occupancy','position','datetime','entity_ts']]

#We want to split the string into only two values and separate them by the comma, while maintaining the - sign in geographical coordinates, especially in the longitude
##The regex expression is made in 3 parts
### (-?\d+\.\d+) returns the longitude and saves the - sign
### \s* separates using the comma, which we will use to create the new two columns
### (-?\d+\.\d+) returns the latitude (and it saves the sign, but as we are in Portugal is not an issue because we will only have positive values here)
df[['longitude', 'latitude']] = df['position'].str.extract(r'\[(-?\d+\.\d+),\s*(-?\d+\.\d+)\]')

#We do not need the original column anymore and so we can drop it, as it is redundant (and wrong)
df = df.drop(columns=['position'])

#This represents the new dataframe with all the columns that we want, in the order that is more suitable in our opinion
df = df[['id', 'id_park', 'name', 'max_capacity', 'occupancy', 'latitude', 'longitude','datetime','entity_ts']]

df.head()
```

```python
#print(df.info()) in order to verify the contents of the dataset
#print(df.duplicated()) in order to see if there are any duplicated values
#dfresult = df.dropna() in order to drop any null values that may exist
##print(dfresult)
#df['column'] = df1['column'].astype('string') if necessary in order to change the values
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='mixed', dayfirst=True)

# Extrair componentes temporais para datetime
# df['month'] = df['datetime'].dt.month -> Não precisamos
#df['day'] = df['datetime'].dt.day -> Não precisamos
df['hour'] = df['datetime'].dt.hour # Poderia ser útil numa análise por hora
df['day_of_week'] = df['datetime'].dt.dayofweek  # Segunda-feira = 0, Domingo = 6
df['per_occupancy'] = (df['occupancy'] / df['max_capacity'] * 100)
df['per_occupancy'] = df['per_occupancy'].round(1)
```

```python
# Por fim devemos retirar as colunas desnecessárias do dataset:
del df['entity_ts']
```

```python
#df_mean = df.groupby(['name', 'day_of_week'])['occupancy'].mean().unstack()

#print(df_mean)

start_date = '2020-01-6' # Podemos alterar para a semana que quisermos assim
end_date = '2020-01-13'
df_week = df[(df['datetime'] >= start_date) & (df['datetime'] <= end_date)]


df_week_mean = df_week.groupby(['name', 'day_of_week'])['occupancy'].mean().unstack()

# Ver o DataFrame com as médias
print(df_week_mean)
```

```python
# Preparar os dados para o heatmap
fig, ax = plt.subplots()
cax = ax.matshow(df_week_mean, cmap='coolwarm')

# Adicionar barra de cores
fig.colorbar(cax)

# Definir os ticks nos eixos
ax.set_xticks(np.arange(len(df_week_mean.columns)))
ax.set_yticks(np.arange(len(df_week_mean.index)))

# Adicionar labels aos ticks
ax.set_xticklabels(df_week_mean.columns)
ax.set_yticklabels(df_week_mean.index)

plt.setp(ax.get_xticklabels(), rotation=0, ha="left",
         rotation_mode="anchor")
        
# Adicionar labels aos eixos
plt.xlabel('Dias da Semana')
plt.ylabel('Parques de Estacionamento')

# Mostrar o heatmap
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Preparing the data for a heatmap
## If we do not format strings, the graph will assume a value that will not fit each cell, specially in parks with high medians
plt.figure(figsize=(8, 9)) 
ax = sns.heatmap(df_week_mean, cmap='OrRd', annot=True, fmt=".1f", linewidth=.5)
ax.set_xticks(np.arange(len(df_week_mean.columns)) + 1)
ax.set_yticks(np.arange(len(df_week_mean.index)) + 0.5)
ax.set_xticklabels(df_week_mean.columns)
ax.set_yticklabels(df_week_mean.index)

plt.show()
```

```python
# Criar o gráfico de linhas
plt.figure(figsize=(20, 10))

for parque in df_week_mean.index:
    plt.plot(df_week_mean.columns, df_week_mean.loc[parque], marker='o', label=parque)

# Adicionar título e rótulos
plt.title('Média de Ocupação por Dia da Semana para Cada Parque')
plt.xlabel('Dia da Semana')
plt.ylabel('Média de Ocupação (%)')

# Adicionar legenda
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)

# Mostrar o gráfico
plt.grid(False)
plt.show()
```
